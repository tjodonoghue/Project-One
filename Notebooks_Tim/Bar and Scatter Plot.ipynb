```python
#We are going to create a bar graph that represents mean price per state of a used car. 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 

file_path = r'C:\Users\Gamer PC\Project-1\Cleaned_Data.csv'  # Replace with your file path
data = pd.read_csv(file_path)
data.head()
```

```python
#First we need to find the mean price of a used car in each state 
mean_price_per_state = data.groupby('State')['Price Adjusted'].mean().sort_values()
mean_price_per_state.head()
```

```python
#Now we set the parameters for our bar graph 
plt.figure(figsize = (20, 15))
color_list = ['forestgreen', 'skyblue', 'orange'] 
plt.bar(mean_price_per_state.index, mean_price_per_state.values, color = color_list)

#Create labels for the graph 
plt.xlabel('State')
plt.ylabel('Average Sale Price')
plt.title('Average Sale Price of a Used Car by State')
plt.xticks(rotation = 45)
plt.show()
```
