```python
#We are going to create a bar graph that represents mean price per state of a used car. 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 

file_path = r'C:\Users\Gamer PC\Project-1\Cleaned_Data.csv'  # Replace with your file path
data = pd.read_csv(file_path)
data.head()
```

```python
#First we need to find the mean price of a used car in each state 
mean_price_per_state = data.groupby('State')['Price'].mean().sort_values()
mean_price_per_state.head()
```

```python
#Now we set the parameters for our bar graph 
plt.figure(figsize = (20, 15))
color_list = ['forestgreen', 'skyblue', 'orange'] 
plt.bar(mean_price_per_state.index, mean_price_per_state.values, color = color_list)

#Create labels for the graph 
plt.xlabel('State')
plt.ylabel('Average Sale Price')
plt.title('Average Sale Price of a Used Car by State')
plt.xticks(rotation = 45)
plt.show()
```

```python
#Next, we'll built a scatter plot showing the relationship between a car's mileage and sales price. This is a fairly simple process- 

#First I am going to add a column to our dataframe that formats the price column with how I would like for it to be displayed on the graph 
data['Formatted Price'] = data['Price'].apply(lambda x: '${:,.2f}'.format(x))

#now we set parameters for the graph
plt.figure(figsize =(12,8))
plt.scatter(data['Formatted Price'], data['Mileage'], alpha = .7, s= 25, color = 'limegreen')

plt.xlabel('Mileage')
plt.ylabel('Price')
plt.title('Relationship Between Milage and Price')


plt.show()
```
